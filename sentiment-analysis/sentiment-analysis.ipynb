```python
import torch
import transformers

transformers.__version__
```

```python
from transformers import pipeline 
sentiment_pipeline = pipeline("sentiment-analysis") 
data = ["I love you", 
        "I hate you", 
        "I might hate for this but I know its good for me", 
        "have you tried this game",
        "maybe i will do it later",
        "since you didnt read the manual i had to try it myself",
        "You had no idea what it is about"]
sentiment_pipeline(data)
```

We can always use a specific sentiment analysis model for our needs or language.
For eg for tweets 

```python
specific_model = pipeline(model="finiteautomata/bertweet-base-sentiment-analysis") 
```

```python
specific_model(data)
```
